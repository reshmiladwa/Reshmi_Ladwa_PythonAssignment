```python
#Determine service setting that reported the most appointments in North West London from 1 January to 1 June 2022
nwlondon=nc.loc[nc['sub_icb_location_name'].str.contains("NHS North West London ICB - W2U3Z",case=False) &(nc['appointment_date']>= '2022-01-01') &\
               (nc['appointment_date']< '2022-06-02')]

nwlondon.nlargest(n=5, columns=['count_of_appointments'],keep='all') 
```

```python
#Count of appointments per month 
monthly_apps = nc.groupby('appointment_month').sum()
monthly_apps.sort_values(by=['count_of_appointments'], inplace=True)
print(monthly_apps)
```

```python
#Dataframe with only text values
tweet_text=tweets['tweet_full_text']

tweet_text
```

```python
tags=[]
```

```python
#Loop through the messages to createa list of values containing hashtags
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
            
#Display all tags             
tags
```

```python
hashtags=pd.Series(tags).value_counts()
```
